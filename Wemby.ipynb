```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/drive/MyDrive/NBA_ML_Project/WembyStats.csv")
```

```python
df.head()
```

```python
df = df[df['GS'] != 'Inactive']
df.columns
```

```python
wembyDF = df[['Date', 'PTS']]
wembyDF.head()
```

```python
wembyDF['Date'] = pd.to_datetime(wembyDF['Date'])
```

```python
wembyDF.info()
wembyDF.head()
```

```python
wembyDF.sort_values('Date')
```

```python
wembyDF['Last3'] = wembyDF.rolling(window=3, min_periods=1).PTS.mean()
wembyDF['Last5'] = wembyDF.rolling(window=5, min_periods=1).PTS.mean()
wembyDF['Last7'] = wembyDF.rolling(window=7, min_periods=1).PTS.mean()
```

```python
wembyDF.head()
```

```python
model = DecisionTreeClassifier(random_state=0)
X = wembyDF[['Last3', 'Last5', 'Last7']]
y = wembyDF['PTS']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
fitted_model = model.fit(X_train, y_train)
fitted_model.score(X_test, y_test)
```

```python
predictions = fitted_model.predict(X_test)
```
