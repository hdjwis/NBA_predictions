```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
team_stats = pd.read_csv('/content/drive/MyDrive/NBA_ML_Project/NBA_team_ratings.csv')
wemby_stats = pd.read_csv('/content/drive/MyDrive/NBA_ML_Project/WembyStats.csv')
```

```python
team_stats.head(30)
```

```python
team_stats['Team'].replace(inplace=True, to_replace={
    'Oklahoma City Thunder' : 'OKC',
    'Cleveland Cavaliers' : 'CLE',
    'Boston Celtics' : 'BOS',
    'Memphis Grizzlies' : 'MEM',
    'Denver Nuggets' : 'DEN',
    'Houston Rockets' : 'HOU',
    'New York Knicks' : 'NYK',
    'Minnesota Timberwolves' : 'MIN',
    'Dallas Mavericks' : 'DAL',
    'Los Angeles Clippers' : 'LAC',
    'Golden State Warriors' : 'GSW',
    'Los Angeles Lakers' : 'LAL',
    'Sacramento Kings' : 'SAC',
    'Milwaukee Bucks' : 'MIL',
    'Indiana Pacers' : 'IND',
    'Detroit Pistons' : 'DET',
    'San Antonio Spurs' : 'SAS',
    'Miami Heat' : 'MIA',
    'Orlando Magic' : 'ORL',
    'Phoenix Suns' : 'PHO',
    'Atlanta Hawks' : 'ATL',
    'Philadelphia 76ers' : 'PHI',
    'Chicago Bulls' : 'CHI',
    'Toronto Raptors' : 'TOR',
    'Portland Trail Blazers' : 'POR',
    'Charlotte Hornets' : 'CHO',
    'Brooklyn Nets' : 'BRK',
    'Utah Jazz' : 'UTA',
    'New Orleans Pelicans' : 'NOP',
    'Washington Wizards' : 'WAS'
})
```

```python
team_stats.head()
```

```python
team_stats.drop(columns=['Rk', 'Conf', 'Div', 'W', 'L'], inplace=True)
```

```python
wemby_stats.rename(columns={'Tm':'Team'}, inplace=True)
```

```python
wemby_stats['W/L%'] = 0.0
wemby_stats['MOV'] = 0.0
wemby_stats['ORtg'] = 0.0
wemby_stats['DRtg'] = 0.0
wemby_stats['NRtg'] = 0.0
wemby_stats['MOV/A'] = 0.0
wemby_stats['ORtg/A'] = 0.0
wemby_stats['DRtg/A'] = 0.0
wemby_stats['NRtg/A'] = 0.0
```

```python
wemby_stats.head(30)
```

```python
for index, row in wemby_stats.iterrows():
  wemby_stats.loc[index, 'W/L%'] = team_stats.loc[team_stats.Team == row.Opp]['W/L%'].values[0]
  wemby_stats.loc[index, 'MOV'] = team_stats.loc[team_stats.Team == row.Opp]['MOV'].values[0]
  wemby_stats.loc[index, 'ORtg'] = team_stats.loc[team_stats.Team == row.Opp]['ORtg'].values[0]
  wemby_stats.loc[index, 'DRtg'] = team_stats.loc[team_stats.Team == row.Opp]['DRtg'].values[0]
  wemby_stats.loc[index, 'NRtg'] = team_stats.loc[team_stats.Team == row.Opp]['NRtg'].values[0]
  wemby_stats.loc[index, 'MOV/A'] = team_stats.loc[team_stats.Team == row.Opp]['MOV/A'].values[0]
  wemby_stats.loc[index, 'ORtg/A'] = team_stats.loc[team_stats.Team == row.Opp]['ORtg/A'].values[0]
  wemby_stats.loc[index, 'DRtg/A'] = team_stats.loc[team_stats.Team == row.Opp]['DRtg/A'].values[0]
  wemby_stats.loc[index, 'NRtg/A'] = team_stats.loc[team_stats.Team == row.Opp]['NRtg/A'].values[0]
```

```python
wemby_stats.head(100)
```

```python
wemby_stats = wemby_stats[wemby_stats['GS'] != 'Inactive']
```

```python
wembyDF = wemby_stats[['Date', 'PTS', 'W/L%',	'MOV',	'ORtg',	'DRtg',	'NRtg',	'MOV/A',	'ORtg/A',	'DRtg/A',	'NRtg/A']]
```

```python
wembyDF['Date'] = pd.to_datetime(wembyDF['Date'])
```

```python
wembyDF.sort_values('Date')
```

```python
wembyDF['PTS'] = wembyDF['PTS'].astype(float)
wembyDF['Last3'] = wembyDF.rolling(window=3, min_periods=1).PTS.mean()
wembyDF['Last5'] = wembyDF.rolling(window=5, min_periods=1).PTS.mean()
wembyDF['Last7'] = wembyDF.rolling(window=7, min_periods=1).PTS.mean()
```

```python
wembyDF.head()
```

```python
X = wembyDF[['W/L%', 'MOV',	'ORtg',	'DRtg',	'NRtg',	'MOV/A',	'ORtg/A',	'DRtg/A',	'NRtg/A',	'Last3',	'Last5',	'Last7']]
y = wembyDF['PTS']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
regression_predictions = regression_model.predict(X_test)
regression_testDF = pd.DataFrame(X_test, columns=['W/L%', 'MOV',	'ORtg',	'DRtg',	'NRtg',	'MOV/A',	'ORtg/A',	'DRtg/A',	'NRtg/A',	'Last3',	'Last5',	'Last7'])
regression_testDF['actual'] = y_test
regression_testDF['predicted'] = regression_predictions
regression_testDF['error'] = abs(regression_testDF['actual'] - regression_testDF['predicted'])
```

```python
regression_testDF.describe()
```

```python
old_regression_model = LinearRegression()
old_regression_model.fit(X_train[['Last3', 'Last5', 'Last7']], y_train)
old_regression_predictions = old_regression_model.predict(X_test[['Last3', 'Last5', 'Last7']])
old_regression_testDF = pd.DataFrame(X_test, columns=['Last3',	'Last5',	'Last7'])
old_regression_testDF['actual'] = y_test
old_regression_testDF['predicted'] = old_regression_predictions
old_regression_testDF['error'] = abs(old_regression_testDF['actual'] - old_regression_testDF['predicted'])
```

```python
old_regression_testDF.describe()
```

```python
rating_regression_model = LinearRegression()
rating_regression_model.fit(X_train[['Last3', 'Last5', 'Last7', 'W/L%',	'MOV',	'ORtg',	'DRtg',	'NRtg']], y_train)
rating_regression_predictions = rating_regression_model.predict(X_test[['Last3', 'Last5', 'Last7', 'W/L%',	'MOV',	'ORtg',	'DRtg',	'NRtg']])
rating_regression_testDF = pd.DataFrame(X_test, columns=['W/L%', 'MOV',	'ORtg',	'DRtg',	'NRtg',	'Last3',	'Last5',	'Last7'])
rating_regression_testDF['actual'] = y_test
rating_regression_testDF['predicted'] = rating_regression_predictions
rating_regression_testDF['error'] = abs(rating_regression_testDF['actual'] - rating_regression_testDF['predicted'])
rating_regression_testDF.describe()
```
